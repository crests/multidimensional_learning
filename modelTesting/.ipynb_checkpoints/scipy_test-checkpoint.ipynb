```python
def base_func(x,a,b,c):
    y = c + (x - b)**2 + (x-a)**3
    return y
def plot_fig(x,y,para):
    fit = base_func(x,para[0],para[1],para[2])
    r2 = r2_score(y, fit)
    print()
    print('R2:',r2)
    print('Results:', para)
    plt.scatter(x,y,label='data')
    plt.plot(x,fit,'r',label='fit')
    plt.legend()
    plt.show()
    return

# fitするデータを作成
x = np.arange(-30, 100, 1)
para = [2.0,5.0,10.0]
np.random.seed(seed=5)
y = base_func(x,para[0],para[1],para[2])+np.random.normal(0, 60, len(x))

#カーブフィッティング
cf = optimize.curve_fit(f=base_func, xdata=x, ydata=y, p0=para)
print(cf)
print("init_param: {}".format(para))
print("optimized_param: {}".format(cf[0]))
#print(cf[0])
plot_fig(x,y,cf[0])
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import rosen, differential_evolution, minimize, basinhopping
from mpl_toolkits.mplot3d import Axes3D

def simple_fitness(x):
    print("{},{},{},{}".format(x[0],x[1],x[2],x[3]))
    return (1.5-x[0]+x[0]*x[1])**2+(2.25-x[0]+x[0]*x[1]**2)**2+(2.625-x[0]+x[0]*x[1]**3)**2

bounds = [(-4.5, 4.5), (-4.5, 4.5)]

#res = differential_evolution(simple_fitness, bounds)
res = basinhopping(simple_fitness, bounds)
print(res)

fig = plt.figure()
ax = fig.gca(projection='3d')

#x = np.arange(-4.5, 4.5, 0.4)
#y = np.arange(-4.5, 4.5, 0.4)
x = np.arange(0,4,0.4)
y = np.arange(0,4,0.4)
xx, yy = np.meshgrid(x, y, sparse=True)
z = (1.5-xx+xx*yy)**2+(2.25-xx+xx*yy**2)**2+(2.625-xx+xx*yy**3)**2

ax.plot_wireframe(xx, yy, z)
ax.scatter(res.x[0], res.x[1], res.fun, s=100, c="r")
```
